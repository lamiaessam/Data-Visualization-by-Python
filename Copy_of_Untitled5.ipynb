```python
Countries = ['Egypt', 'Saudi Arabia', 'UK', 'Germany', 'Syria']
Sales = [1000, 3000, 4000, 5000, 6000]
Profit = [100, 200, 300, 400, 500]


width = 0.35
x = np.arange(len(Countries))



plt.bar(x - width/2, Sales, width,
        color='skyblue', edgecolor='black', linewidth=2, hatch='//')

plt.bar(x + width/2, Profit, width,
        color='red', edgecolor='black', linewidth=2, hatch='//')


plt.xlabel('Countries')
plt.ylabel('Sales')
plt.title('Sales and Profit')
plt.xticks(x, Countries)


for i, S in enumerate(Sales):
    plt.text(i - width/2, S + 0.1, str(S), ha='center')

for i, S in enumerate(Profit):
    plt.text(i + width/2, S + 0.1, str(S), ha='center')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Countries = ['Egypt', 'Saudi Arabia', 'UK', 'Germany', 'Syria']
Sales = [1000, 2000, 3000, 4000, 5000]
Profit = [100, 200, 300, 400, 500]

x_pos = np.arange(len(Countries))

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(x_pos, Sales, color='skyblue', label='Sales', alpha=0.7)
ax1.set_xlabel('Countries')
ax1.set_ylabel('Sales', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax1.set_xticks(x_pos)
ax1.set_xticklabels(Countries, rotation=45, ha='right')

ax2 = ax1.twinx()

ax2.plot(x_pos, Profit, color='red', marker='o', linestyle='-.', label='Profit')
ax2.set_ylabel('Profit', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Sales and Profit by Country')

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()
```

```python
### Draw Chart To Track Down Home Expenses ###
```

```python
exp_vals = [1400 ,600 ,300 ,250 , 100]
exp_labels = ["Home Rent" ,"food" ,"Phone/Internet", "CAR", "Other visities"]
plt.pie(exp_vals ,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals ,labels=exp_labels ,shadow=True)
plt.axis('equal')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels=exp_labels ,shadow=True , autopct='%1.3f%%' , radius=1.66)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels=exp_labels ,shadow=True , autopct='%1.3f%%' , radius=1.66 , explode=(.2 ,0 ,0.2 ,0 ,0) , counterclock=True)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals , labels=exp_labels ,shadow=True , autopct='%1.3f%%' , radius=1.66 , startangle=180 , explode=(.2 ,0 ,0 ,0 ,0) , counterclock=True)
plt.show()
```
